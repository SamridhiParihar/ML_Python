#MATPLOTLIB


> Useful for plotting the datas



```python
#import matplotlib library
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x=np.linspace(0,10,100)
y=np.sin(x)
z=np.cos(x)
```

```python
print(x)
```

```python
print(y)
```

#plotting the data

```python
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,z)
plt.show()
```

```python
plt.plot(z,y)
plt.show()
```

```python
#adding title to x axis and y axis label
plt.plot(x,y)
plt.title('sine wave')
plt.xlabel('angle')
plt.ylabel('sine value')
plt.show()
```

```python
#plot a parabola
x=np.linspace(-10,10,20)
y=x**2
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y,'r+') # printing with plus symbol and red colours
plt.show()
```

```python
plt.plot(x,y,'g.')
plt.show()
```

```python
#multiple lines in a single graph
x=np.linspace(-5,5,50)
plt.plot(x,np.sin(x),'g-')
plt.plot(x,np.cos(x),'r--')
plt.show()
```

#Bar Plots

```python
fig=plt.figure()#creating an empty figure
ax=fig.add_axes([0,0,1,1]) # area to rectangle(enclose the plot for .plot() we don't require it) : 0,0 is origin and 1,1 height and width and ax represents axis()
languages=["eng","hindi","spanish","german","french"]
people=[100,89,90,34,56]
ax.bar(languages,people)
plt.xlabel('languages')
plt.ylabel('no of people speaking')
plt.show()
```

#PIE CHART



```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1]) # area to rectangle(enclose the plot for .plot() we don't require it) : 0,0 is origin and 1,1 height and width and ax represents axis()
languages=["eng","hindi","spanish","german","french"]
people=[100,89,90,34,56]
ax.pie(people,labels=languages ,autopct='%1.1f%%')
plt.show()
```

#Scatter Plot


```python
x=np.linspace(0,10,30)
y=np.sin(x)
z=np.cos(x)
fig2=plt.figure()
ax=fig2.add_axes([0,0,1,1])
ax.scatter(x,y,color='g')
ax.scatter(x,z,color='b')
plt.show()
```

#3D Scatter Plot

```python
fig3=plt.figure()
ax=plt.axes(projection='3d')
z=20*np.random.random(100)
x=np.sin(z)
y=np.cos(z)
ax.scatter(x,y,z,c=z,cmap='Blues')
plt.show()
```
